```python
#W6A1
nums = list(map(int, input().split()))
result = []
for x in nums:
    if x not in result:
       result.append(x)
print(result)
```

```python
#W6A2
nums=list(map(int,input().split()))
result=[]
a=0
for i in nums:
  a+=i
  result.append(a)
print(result)
```

```python
#W6A5:
nums=list(map(int,input().split()))
pos=neg=zero=0
for i in nums:
  if i>0:
    pos+=1
  if i<0:
    neg+=1
  if i==0:
    zero+=1
result={'Positive':pos,'Negative':neg,'Zero':zero}
print(result)
```

```python
#W6A8:
text=input().split()
counts={}
for i in text:
  if i in counts:
    counts[i]+=1
  else:
    counts[i]=1
print(counts)
```
