```python
#W8A1:
a,b=map(int,input().split())
try:
  ket_qua=a/b
  print(f"{ket_qua:.2f}")
except ZeroDivisionError:
  print("Loi chia khong")
```

```python
#W8A2:
try:
  n=int(input())
  my_list=list(map(int,input().split()))
  print(my_list[n])
except IndexError:
  print("IndexError")
```

```python
#W8A4:
try:
  a,b=map(int,input().split())
  print(a+b)
except ValueError:
  print("ValueError")
else:
  print("Khong co loi")
finally:
  print("Ket thuc chuong trinh")
```

```python
#W8A5:
lst=input().split()
result=[]
for x in lst:
  try:
    result.append(int(x))
  except:
    print(f"ValueError, cho {x} cut")
print(*result)
```

```python
#w10a6:
x=float(input())
try:
  if x>=0:
    ket_qua=x**0.5
    print(f"{ket_qua:.2f}")
  else:
    raise ValueError
except ValueError:
  print("So am")
```

```python
try:
    x = int(input())
    if x <= 0:
        raise ValueError
    print(2025 - x)
except ValueError:
    print("Tuoi khong hop le")
```
