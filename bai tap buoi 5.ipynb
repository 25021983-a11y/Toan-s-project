```python
#W5A1
def Max_Of_Two(a, b) -> int:
    if a > b:
        return a
    else:
        return b
print("Kết quả so sánh 23 và -90 là:", Max_Of_Two(23, -90))
print("Kết quả so sánh 234 và 345 là:", Max_Of_Two(234, 345))
print("Kết quả so sánh 1 và 90 là:", Max_Of_Two(1, 90))
print("Kết quả so sánh 13 và 78 là:", Max_Of_Two(13, 78))
```

```python
#W5A3:
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
n = int(input("Nhập n: "))
print("Kết quả kiểm tra số nguyên tố:", is_prime(n))
```

```python
#W5A4
def is_perfect_number(n):
  if n<=0:
     return False
  tong=0
  for i in range(1,n):
     if n % i==0:
      tong+=i
  if tong==n:
    return True
  return False
n=int(input())
print("kiem tra la so hoan hao:", is_perfect_number(n))




```

```python
#W5A6
def tinh_giai_thua(n):
  tich = 1
  if n<0:
    return False
  if n>=0:
   for i in range(1,n+1):
    tich*=i
  return tich
n=int(input())
print("ket qua giai thua:", tinh_giai_thua(n))
```
