```python
#a1
class cylinder:
    def __init__(self,rad,h):
        self.rad=rad
        self.h=h
    def sbm(self):
        return 2*3.14*(self.rad)*(self.rad+self.h)
    def thetich(self):
        return 3.14*((self.rad)**2)*self.h
a,b=map(float,input().split())
c=cylinder(a,b)
print(f"{c.sbm():.2f}")
print(f"{c.thetich():.2f}")
    
```

```python
#a2
class Date():
    def __init__(self,d,m,y):
        self.d=d
        self.m=m
        self.y=y
    def songaytrongthang(self):
      if self.m in [1,3,5,7,8,10,12]:
        return 31
      if self.m in [4,6,9,11]:
        return 30
      if (self.y % 400 == 0) or (self.y % 4 == 0 and self.y % 100 != 0):
        return 29
      return 28
    def tontai(self):
       if self.y<1 or not (1<=self.m<=12):
          return False
       return 1<= self.d <= self.songaytrongthang()
    def nextday(self):
       max_day = self.songaytrongthang()

       if self.d < max_day:
        return (self.d+1, self.m, self.y)
       else:
          if self.m < 12:
            return (1, self.m+1, self.y)
          else:
            return (1, 1, self.y+1)
d,m,y=map(int,input().split())
s=Date(d,m,y)
if s.tontai():
   print(s.nextday())
else:
   print("INVALID")
        



```

```python
#3
class point:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def checkO(self):
        if self.x==0 and self.y==0:
            return True
        return False
    def checkOx(self):
        if self.y==0:
            return True
        return False
    def checkOy(self):
        if self.x==0:
            return True 
        return False
    def distance(self,object):
        a=(self.x-object.x)**2
        b=(self.x-object.x)**2
        return (a+b)**0.5
M=point(0,0)
O=point(0,0)
if M.checkO():
    print("M là gốc tọa độ ")
    print("kcach M đến O",0.00)
elif M.checkOx():
    print("M nằm trên trục Ox")
    print(M.distance(O))
elif M.checkOy():
    print("M nằm trên Oy")
    print(M.distance(O))
else:
    print("M ko nằm trên OX,OY,ko phải gốc tọa độ")
    print(f"{M.distance(O):.2f}")
    

```

```python
#6
class rectangle():
    def __init__(self,w,h):
        self.w=w
        self.h=h
    def area(self):
        return self.w*self.h
    def perimeter(self):
        return 2*(self.w+self.h)
    def scale(self,k):
        self.w*=k
        self.h*=k
w,h,k=map(float,input().split())
hcn=rectangle(w,h)
hcn.scale(k)
print(f"{hcn.area():.2f} {hcn.perimeter():.2f}")

```

```python
#7
class fraction():
    def __init__(self,t,m):
        if m==0:
             raise ValueError("lỗi mẫu số bằng 0")
        elif m<0:
            m=-m
            t=-t
        self.t=t//self.rutgon(abs(t),abs(m))
        self.m=m//self.rutgon(abs(t),abs(m)) 
    def rutgon(self,t1,m1):
        while m1!=0:
          t1,m1=m1,t1%m1
        return t1
    def add(self,other):
        return fraction(self.t*other.m+self.m*other.t,self.m*other.m)
    def sub(self,other):
        return fraction(self.t*other.m-self.m*other.t,self.m*other.m)
    def mul(self,other):
        return fraction(self.t*other.t,self.m*other.m)
    def div(self,other):
        return fraction(self.t*other.m,self.m*other.t)
    def __str__(self):
        return f"{self.t}/{self.m}"
a,b,op,c,d=input().split()
a,b,c,d=int(a),int(b),int(c),int(d)
frac1=fraction(a,b)
frac2=fraction(c,d)
if op=="+":
    res=frac1.add(frac2)
elif op=="-":
    res=frac1.sub(frac2)
elif op=="*":
    res=frac1.mul(frac2)
elif op=="/":
    res=frac1.div(frac2)
print(res)




            

```

```python
#8
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = float(balance)

    def deposit(self, x):
        self.balance += float(x)

    def withdraw(self, x):
        x = float(x)
        if x <= self.balance:
            self.balance -= x

    def __str__(self):
        return f"{self.owner}: {self.balance:.2f}"
first_line = input().split()
owner = first_line[0]
balance = float(first_line[1])

account = BankAccount(owner, balance)

n = int(input())

for _ in range(n):
    command = input().split()
    action = command[0]
    amount = float(command[1])

    if action == "DEPOSIT":
        account.deposit(amount)
    elif action == "WITHDRAW":
        account.withdraw(amount)
print(f"{account.balance:.2f}")
```

```python
#9
class Student:
    def __init__(self, name):
        self.name = name
        self.scores = []

    def add_score(self, subject, score):
        self.scores.append(float(score))

    def avg(self):
        return sum(self.scores) / len(self.scores)

    def rank(self):
        a = self.avg()
        if a >= 8:
            return "Excellent"
        elif a >= 6.5:
            return "Good"
        elif a >= 5:
            return "Average"
        else:
            return "Poor"


name = input()
st = Student(name)

n = int(input())

for _ in range(n):
    sub, sc = input().split()
    st.add_score(sub, sc)

a = st.avg()
print(f"{st.name} {a:.2f} {st.rank()}")
```

```python
#10
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = int(year)


class Library:
    def __init__(self):
        self.books = []

    def add(self, book):
        self.books.append(book)

    def count_by_author(self, name):
        return sum(1 for b in self.books if b.author == name)

    def find_by_year(self, y):
        return sum(1 for b in self.books if b.year == int(y))


lib = Library()

n = int(input())

for _ in range(n):
    line = input()

    if line.startswith("ADD"):
        _, data = line.split()
        title, author, year = data.split(";")
        lib.add(Book(title, author, year))

    elif line.startswith("COUNTYEAR"):
        _, y = line.split()
        print(lib.find_by_year(y))

    elif line.startswith("COUNT"):
        _, author = line.split()
        print(lib.count_by_author(author))
```

```python
#11
import math

def vec(a, b):
    return [b[i] - a[i] for i in range(3)]

def cross(u, v):
    return [
        u[1]*v[2] - u[2]*v[1],
        u[2]*v[0] - u[0]*v[2],
        u[0]*v[1] - u[1]*v[0]
    ]

def dot(u, v):
    return sum(u[i]*v[i] for i in range(3))

def norm(u):
    return math.sqrt(dot(u, u))


A = list(map(float, input().split()))
B = list(map(float, input().split()))
C = list(map(float, input().split()))
D = list(map(float, input().split()))

AB = vec(A, B)
AC = vec(A, C)
BC = vec(B, C)
BD = vec(B, D)

n1 = cross(AB, AC)
n2 = cross(BC, BD)

cos_phi = abs(dot(n1, n2)) / (norm(n1) * norm(n2))
phi = math.degrees(math.acos(cos_phi))

print(f"{phi:.2f}")
```
