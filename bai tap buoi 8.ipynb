```python
#w8a1
class Cylinder:
  def __init__(self,ban_kinh:float,chieu_cao:float):
    self.ban_kinh=ban_kinh
    self.chieu_cao=chieu_cao
  def dien_tich_toan_phan(self):
    return 2*3.14*self.ban_kinh*(self.ban_kinh+self.chieu_cao)
  def the_tich(self):
    return 3.14*(self.ban_kinh)**2*(self.chieu_cao)
r,h=map(float,input().split())
cylinder_obj=Cylinder(r,h)
print(cylinder_obj.dien_tich_toan_phan())
print(cylinder_obj.the_tich())
```

```python
#w8a2:
class Date:
  def __init__(self,d,m,y):
    self.d=d
    self.m=m
    self.y=y
  def nam_nhuan(self):
    return self.y%400==0 or (self.y%4==0 and self.y%100!=0)
  def is_valid(self):
    if self.m<1 or self.m>12:
      return False
    days_in_month=[31,28,31,30,31,30,31,31,30,31,30,31]
    if self.m==2 and self.is_nam_nhuan():
      max_day=29
    else:
      max_day=days_in_month[self.m-1]
    return 1<=self.d<=max_day
  def next_day(self):
    days_in_month=[31,28,31,30,31,30,31,31,30,31,30,31]
    if self.m==2 and self.is_nam_nhuan():
      max_day=29
    else:
      max_day=days_in_month[self.m-1]
    if self.d<max_day:
      self.d+=1
    else:
      self.d=1
      if self.m<12:
        self.m+=1
      else:
        self.m=1
        self.y+=1
  def __str__(self):
    return f"{self.d:02d}/{self.m:02d}/{self.y}"
s=input().strip()
d,m,y=map(int,s.split('/'))
date=Date(d,m,y)
if date.is_valid():
  date.next_day()
  print(date)
else:
  print("Invalid")

```
